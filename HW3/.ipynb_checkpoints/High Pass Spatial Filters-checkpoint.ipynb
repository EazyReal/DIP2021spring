### Laplacian Filter

```python
def exp1(imageName, filterName):
    # Parameters
    ddepth = cv.CV_16S
    kernel_size = 3
    # Load Image
    src = cv.imread(imageName, cv.IMREAD_COLOR) # Load an image
    # Reduce Noise
    src = cv.GaussianBlur(src, (3, 3), 0)
    # Convert the image to grayscale
    src_gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
    # Create Window
    # Apply Laplace function
    dst = cv.Laplacian(src_gray, ddepth, ksize=kernel_size)
    # converting back to uint8
    abs_dst = cv.convertScaleAbs(dst)
    # Display
    plt.imshow(abs_dst, 'gray')
    return abs_dst
```

```python
e1 = exp1('Bodybone.bmp', 'Laplacian')
```

```python
def demo(imageName, filterName):
    # Load Image
    img = cv.imread('Bodybone.bmp', cv.COLOR_BGR2GRAY)
    rgb_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    print(np.array(rgb_img).shape)

    # Convert to Gray
    grayImage = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    # Sobel Operator
    x = cv.Sobel(grayImage, cv.CV_16S, 1, 0)
    y = cv.Sobel(grayImage, cv.CV_16S, 0, 1)

    # Convert to int8, Image Fusion
    absX = cv.convertScaleAbs(x)
    absY = cv.convertScaleAbs(y)
    Sobel = cv.addWeighted(absX, 0.5, absY, 0.5, 0)

    # Display
    titles = ['original', filterName +  operator']
    images = [rgb_img, Sobel]
    for i in range(2):
        plt.subplot(1, 2, i + 1), plt.imshow(images[i], 'gray')
        plt.title(titles[i])
        plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
demo()
```
